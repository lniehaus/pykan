```python
import os
import numpy as np
import pandas as pd


def load_experiment(experiment_path):
    run_paths = [os.path.join(experiment_path, name) for name in os.listdir(experiment_path) if os.path.isdir(os.path.join(experiment_path, name))]

    experiment_df = pd.DataFrame()
    for rp in run_paths:
        #print(rp)


        # load params
        params_df = pd.DataFrame()
        params_path = os.path.join(rp, "params")
        for param_name in os.listdir(params_path):
            param_file = os.path.join(params_path, param_name)
            df = pd.read_csv(param_file, sep=' ', header=None)
            #df.columns = ['Value']
            params_df[param_name] = df[0]
            #run_df[metric_name] = metric_df['Value']

        # Add run_path as identifier for the run
        params_df['run_path'] = rp

        # For metrics, we'll collect the last value of each metric for this run
        metrics_dict = {}
        metrics_path = os.path.join(rp, "metrics") 
        for metric_name in os.listdir(metrics_path):
            metric_file = os.path.join(metrics_path, metric_name)

            df = pd.read_csv(metric_file, sep=' ', header=None)
            df.columns = ['Timestamp', 'Value', 'Index']

            # Take the last value instead of all values
            metrics_dict[metric_name] = df['Value'].iloc[-1]

        # Create a single row DataFrame for this run's metrics
        metrics_df = pd.DataFrame([metrics_dict])

        if not params_df.empty and not metrics_df.empty:
            # Since we now have single values for metrics, we can directly combine with params
            run_df = pd.concat([params_df.iloc[[0]].reset_index(drop=True), metrics_df.reset_index(drop=True)], axis=1)
            
            experiment_df = pd.concat([experiment_df, run_df], axis=0, ignore_index=True)

    return experiment_df
```

```python
# find experiments by name

def find_experiments_by_name(experiment_name_pattern, mlruns_path="../mlruns"):
    """
    Find MLflow experiment IDs based on a name pattern.
    
    Args:
        experiment_name_pattern (str): The base pattern to search for (e.g., "depth_width_grid_42")
        mlruns_path (str): Path to the mlruns directory
    
    Returns:
        dict: Dictionary with keys 'depth', 'width', 'grid' and values as experiment paths
    """
    import os
    import json
    
    # Initialize result dictionary
    experiments = {
        'depth': None,
        'width': None, 
        'grid': None
    }
    
    # Check if mlruns directory exists
    if not os.path.exists(mlruns_path):
        print(f"MLruns directory not found: {mlruns_path}")
        return experiments
    
    # Iterate through all experiment directories in mlruns
    for experiment_id in os.listdir(mlruns_path):
        experiment_dir = os.path.join(mlruns_path, experiment_id)
        
        # Skip if not a directory or if it's the models folder
        if not os.path.isdir(experiment_dir) or experiment_id == "models":
            continue
            
        # Look for meta.yaml file which contains experiment metadata
        meta_file = os.path.join(experiment_dir, "meta.yaml")
        if os.path.exists(meta_file):
            try:
                with open(meta_file, 'r') as f:
                    content = f.read()
                    
                # Look for experiment name in the meta.yaml content
                if f"name: {experiment_name_pattern}_depth" in content:
                    experiments['depth'] = f"{mlruns_path}/{experiment_id}/"
                    print(f"Found depth experiment: {experiment_name_pattern}_depth -> {experiment_id}")
                elif f"name: {experiment_name_pattern}_width" in content:
                    experiments['width'] = f"{mlruns_path}/{experiment_id}/"
                    print(f"Found width experiment: {experiment_name_pattern}_width -> {experiment_id}")
                elif f"name: {experiment_name_pattern}_grid" in content:
                    experiments['grid'] = f"{mlruns_path}/{experiment_id}/"
                    print(f"Found grid experiment: {experiment_name_pattern}_grid -> {experiment_id}")
                    
            except Exception as e:
                # If we can't read the meta.yaml, skip this experiment
                continue
    
    return experiments

def set_experiment_paths(experiment_name_pattern, mlruns_path="../mlruns"):
    """
    Find experiments and set the global experiment path variables.
    
    Args:
        experiment_name_pattern (str): The base pattern to search for (e.g., "depth_width_grid_42")
        mlruns_path (str): Path to the mlruns directory
    
    Returns:
        tuple: (experiment_path_depth, experiment_path_width, experiment_path_grid)
    """
    experiments = find_experiments_by_name(experiment_name_pattern, mlruns_path)
    
    experiment_path_depth = experiments['depth']
    experiment_path_width = experiments['width'] 
    experiment_path_grid = experiments['grid']
    
    print(f"\nExperiment paths for '{experiment_name_pattern}':")
    print(f"# {experiment_name_pattern}_depth")
    print(f"experiment_path_depth = \"{experiment_path_depth}\"")
    print(f"# {experiment_name_pattern}_width")
    print(f"experiment_path_width = \"{experiment_path_width}\"")
    print(f"# {experiment_name_pattern}_grid")
    print(f"experiment_path_grid = \"{experiment_path_grid}\"")
    
    return experiment_path_depth, experiment_path_width, experiment_path_grid

# Example usage:
# experiment_path_depth, experiment_path_width, experiment_path_grid = set_experiment_paths("depth_width_grid_42")

experiment_path_depth, experiment_path_width, experiment_path_grid = set_experiment_paths("depth_width_grid_64")
```

```python
# Display the configuration of the first row in df_depth in a nice format

def display_configuration(df, row_index=0, title="Configuration Details"):
    """
    Display a nicely formatted configuration for a specific row in the dataframe.
    
    Args:
        df (pd.DataFrame): The dataframe containing experiment data
        row_index (int): The index of the row to display (default: 0 for first row)
        title (str): Title for the display
    """
    if df.empty:
        print("‚ö†Ô∏è DataFrame is empty!")
        return
        
    if row_index >= len(df):
        print(f"‚ö†Ô∏è Row index {row_index} is out of bounds! DataFrame has {len(df)} rows.")
        return
    
    row = df.iloc[row_index]
    
    print("=" * 70)
    print(f"üîß {title} (Row {row_index + 1})")
    print("=" * 70)

    # Network Architecture Parameters
    print("\nüìê NETWORK ARCHITECTURE")
    print("-" * 30)
    arch_params = ['hidden_depth', 'hidden_width', 'grid', 'k']
    for param in arch_params:
        if param in row.index:
            print(f"  ‚Ä¢ {param.replace('_', ' ').title():<15}: {row[param]}")
    
    # Dataset Information
    print("\nüìä DATASET INFORMATION")
    print("-" * 30)
    dataset_params = ['dataset', 'dataset_name', 'input_dim', 'output_dim', 'n_samples', 'train_size', 'test_size']
    found_dataset_params = False
    for param in dataset_params:
        if param in row.index:
            found_dataset_params = True
            value = row[param]
            param_name = param.replace('_', ' ').title()
            if isinstance(value, str):
                print(f"  ‚Ä¢ {param_name:<15}: {value}")
            elif isinstance(value, (int, float)):
                if 'size' in param or 'samples' in param or 'dim' in param:
                    print(f"  ‚Ä¢ {param_name:<15}: {int(value):,}")
                else:
                    print(f"  ‚Ä¢ {param_name:<15}: {value}")
    
    if not found_dataset_params:
        print("  ‚Ä¢ No dataset information available")
    
    # Training Parameters
    print("\nüéØ TRAINING PARAMETERS")
    print("-" * 30)
    train_params = ['epochs', 'lr', 'batch_size', 'seed']
    for param in train_params:
        if param in row.index:
            value = row[param]
            if param == 'lr':
                print(f"  ‚Ä¢ Learning Rate    : {value:.6f}")
            else:
                print(f"  ‚Ä¢ {param.replace('_', ' ').title():<15}: {value}")
    
    # Model Complexity
    print("\nüìä MODEL COMPLEXITY")
    print("-" * 30)
    complexity_params = ['total_parameters', 'total_flops']
    for param in complexity_params:
        if param in row.index:
            value = row[param]
            if 'parameters' in param:
                print(f"  ‚Ä¢ Total Parameters : {int(value):,}")
            elif 'flops' in param:
                print(f"  ‚Ä¢ Total FLOPs      : {int(value):,}")
    
    # Performance Metrics
    print("\nüéØ PERFORMANCE METRICS")
    print("-" * 30)
    metrics = ['train_acc', 'test_acc', 'train_loss', 'test_loss']
    for metric in metrics:
        if metric in row.index:
            value = row[metric]
            metric_name = metric.replace('_', ' ').title()
            if 'acc' in metric:
                print(f"  ‚Ä¢ {metric_name:<15}: {value:.4f} ({value*100:.2f}%)")
            else:
                print(f"  ‚Ä¢ {metric_name:<15}: {value:.6f}")
    
    # Additional Parameters (catch all others)
    print("\n‚öôÔ∏è OTHER PARAMETERS")
    print("-" * 30)
    displayed_params = {'hidden_depth', 'hidden_width', 'grid', 'k', 'epochs', 'lr', 
                       'batch_size', 'seed', 'total_parameters', 'total_flops',
                       'train_acc', 'test_acc', 'train_loss', 'test_loss', 'run_path',
                       'dataset', 'dataset_name', 'input_dim', 'output_dim', 'n_samples', 'train_size', 'test_size'}
    
    other_params = [col for col in row.index if col not in displayed_params]
    
    if other_params:
        for param in sorted(other_params):
            value = row[param]
            param_name = param.replace('_', ' ').title()
            if isinstance(value, float):
                print(f"  ‚Ä¢ {param_name:<15}: {value:.6f}")
            else:
                print(f"  ‚Ä¢ {param_name:<15}: {value}")
    else:
        print("  ‚Ä¢ No additional parameters")
    
    # Run Information
    if 'run_path' in row.index:
        print("\nüìÅ RUN INFORMATION")
        print("-" * 30)
        run_path = str(row['run_path'])
        run_id = run_path.split('/')[-1] if '/' in run_path else run_path
        print(f"  ‚Ä¢ Run ID           : {run_id}")
        print(f"  ‚Ä¢ Full Path        : {run_path}")
    
    print("\n" + "=" * 70)

# Display the first row configuration
if not df_depth.empty:
    display_configuration(df_depth, row_index=0, title="Depth Experiment - First Configuration")
else:
    print("‚ö†Ô∏è df_depth is empty! Please load the experiment data first.")
```

```python
# Comprehensive Analysis and Comparison
print("="*80)
print("COMPREHENSIVE EXPERIMENTAL ANALYSIS")
print("="*80)

# Summary statistics for each experiment
experiments = {
    'Depth': df_depth,
    'Width': df_width, 
    'Grid': df_grid
}

print("\n1. SUMMARY STATISTICS")
print("-"*50)
for name, df in experiments.items():
    print(f"\n{name} Experiment:")
    print(f"  ‚Ä¢ Sample size: {len(df)} configurations")
    print(f"  ‚Ä¢ Train Accuracy: {df['train_acc'].mean():.3f} ¬± {df['train_acc'].std():.3f} (range: {df['train_acc'].min():.3f} - {df['train_acc'].max():.3f})")
    print(f"  ‚Ä¢ Test Accuracy:  {df['test_acc'].mean():.3f} ¬± {df['test_acc'].std():.3f} (range: {df['test_acc'].min():.3f} - {df['test_acc'].max():.3f})")
    print(f"  ‚Ä¢ Total Parameters: {df['total_parameters'].mean():.0f} ¬± {df['total_parameters'].std():.0f} (range: {df['total_parameters'].min():.0f} - {df['total_parameters'].max():.0f})")
    print(f"  ‚Ä¢ Total FLOPs: {df['total_flops'].mean():.0f} ¬± {df['total_flops'].std():.0f} (range: {df['total_flops'].min():.0f} - {df['total_flops'].max():.0f})")

print("\n\n2. BEST PERFORMING CONFIGURATIONS")
print("-"*50)
for name, df in experiments.items():
    best_train_idx = df['train_acc'].idxmax()
    best_test_idx = df['test_acc'].idxmax()
    best_train_row = df.loc[best_train_idx]
    best_test_row = df.loc[best_test_idx]
    
    print(f"\n{name} Experiment - Best Train Accuracy ({best_train_row['train_acc']:.3f}):")
    print(f"  ‚Ä¢ Configuration: depth={best_train_row['hidden_depth']}, width={best_train_row['hidden_width']}, grid={best_train_row['grid']}")
    print(f"  ‚Ä¢ Parameters: {best_train_row['total_parameters']:.0f}, FLOPs: {best_train_row['total_flops']:.0f}")
    print(f"  ‚Ä¢ Test accuracy: {best_train_row['test_acc']:.3f}")
    
    print(f"\n{name} Experiment - Best Test Accuracy ({best_test_row['test_acc']:.3f}):")
    print(f"  ‚Ä¢ Configuration: depth={best_test_row['hidden_depth']}, width={best_test_row['hidden_width']}, grid={best_test_row['grid']}")
    print(f"  ‚Ä¢ Parameters: {best_test_row['total_parameters']:.0f}, FLOPs: {best_test_row['total_flops']:.0f}")
    print(f"  ‚Ä¢ Train accuracy: {best_test_row['train_acc']:.3f}")

print("\n\n3. EFFICIENCY ANALYSIS")
print("-"*50)
# Calculate efficiency metrics: accuracy per parameter and accuracy per FLOP
efficiency_results = {}
for name, df in experiments.items():
    df_copy = df.copy()
    df_copy['train_acc_per_param'] = df_copy['train_acc'] / df_copy['total_parameters']
    df_copy['test_acc_per_param'] = df_copy['test_acc'] / df_copy['total_parameters']
    df_copy['train_acc_per_flop'] = df_copy['train_acc'] / df_copy['total_flops']
    df_copy['test_acc_per_flop'] = df_copy['test_acc'] / df_copy['total_flops']
    
    efficiency_results[name] = df_copy
    
    best_train_eff_param_idx = df_copy['train_acc_per_param'].idxmax()
    best_test_eff_param_idx = df_copy['test_acc_per_param'].idxmax()
    best_train_eff_flop_idx = df_copy['train_acc_per_flop'].idxmax()
    best_test_eff_flop_idx = df_copy['test_acc_per_flop'].idxmax()
    
    print(f"\n{name} Experiment:")
    print(f"  ‚Ä¢ Most efficient (train acc/param): {df_copy.loc[best_train_eff_param_idx, 'train_acc_per_param']:.6f}")
    print(f"    Configuration: depth={df_copy.loc[best_train_eff_param_idx, 'hidden_depth']}, width={df_copy.loc[best_train_eff_param_idx, 'hidden_width']}, grid={df_copy.loc[best_train_eff_param_idx, 'grid']}")
    print(f"  ‚Ä¢ Most efficient (test acc/param): {df_copy.loc[best_test_eff_param_idx, 'test_acc_per_param']:.6f}")
    print(f"    Configuration: depth={df_copy.loc[best_test_eff_param_idx, 'hidden_depth']}, width={df_copy.loc[best_test_eff_param_idx, 'hidden_width']}, grid={df_copy.loc[best_test_eff_param_idx, 'grid']}")
    print(f"  ‚Ä¢ Most efficient (train acc/FLOP): {df_copy.loc[best_train_eff_flop_idx, 'train_acc_per_flop']:.8f}")
    print(f"    Configuration: depth={df_copy.loc[best_train_eff_flop_idx, 'hidden_depth']}, width={df_copy.loc[best_train_eff_flop_idx, 'hidden_width']}, grid={df_copy.loc[best_train_eff_flop_idx, 'grid']}")
    print(f"  ‚Ä¢ Most efficient (test acc/FLOP): {df_copy.loc[best_test_eff_flop_idx, 'test_acc_per_flop']:.8f}")
    print(f"    Configuration: depth={df_copy.loc[best_test_eff_flop_idx, 'hidden_depth']}, width={df_copy.loc[best_test_eff_flop_idx, 'hidden_width']}, grid={df_copy.loc[best_test_eff_flop_idx, 'grid']}")

print("\n\n4. COMPARATIVE EFFICIENCY RANKINGS")
print("-"*50)
# Find overall most efficient configurations across all experiments
all_configs = []
for name, df in efficiency_results.items():
    for idx, row in df.iterrows():
        all_configs.append({
            'experiment': name,
            'config': f"depth={row['hidden_depth']}, width={row['hidden_width']}, grid={row['grid']}",
            'train_acc': row['train_acc'],
            'test_acc': row['test_acc'], 
            'total_parameters': row['total_parameters'],
            'total_flops': row['total_flops'],
            'train_acc_per_param': row['train_acc_per_param'],
            'test_acc_per_param': row['test_acc_per_param'],
            'train_acc_per_flop': row['train_acc_per_flop'],
            'test_acc_per_flop': row['test_acc_per_flop']
        })

all_configs_df = pd.DataFrame(all_configs)

print("\nTop 3 Most Parameter-Efficient (Test Accuracy/Parameter):")
top_param_eff = all_configs_df.nlargest(3, 'test_acc_per_param')
for i, (idx, row) in enumerate(top_param_eff.iterrows(), 1):
    print(f"  {i}. {row['experiment']} - {row['config']}")
    print(f"     Test Acc: {row['test_acc']:.3f}, Parameters: {row['total_parameters']:.0f}, Efficiency: {row['test_acc_per_param']:.6f}")

print("\nTop 3 Most FLOP-Efficient (Test Accuracy/FLOP):")
top_flop_eff = all_configs_df.nlargest(3, 'test_acc_per_flop')
for i, (idx, row) in enumerate(top_flop_eff.iterrows(), 1):
    print(f"  {i}. {row['experiment']} - {row['config']}")
    print(f"     Test Acc: {row['test_acc']:.3f}, FLOPs: {row['total_flops']:.0f}, Efficiency: {row['test_acc_per_flop']:.8f}")
```
