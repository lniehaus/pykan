```python
import numpy as np

def radial_basis_function(x, c, beta):
    """
    Calculate the Radial Basis Function (RBF).
    
    Parameters:
        x (numpy.ndarray): Input data point(s).
        c (numpy.ndarray): Center of the RBF.
        beta (float): Spread parameter of the RBF.
        
    Returns:
        numpy.ndarray: RBF value(s).
    """
    return np.exp(-beta * np.linalg.norm(x - c, axis=-1)**2)

# Example usage
x = np.array([1.0, 2.0])
c = np.array([0.0, 0.0])
#beta = 1.0
beta = 0.33

result = radial_basis_function(x, c, beta)
print("RBF value:", result)
```

```python
import matplotlib.pyplot as plt
# Generate a range of x values for plotting
x_values = np.linspace(-2, 2, 100)
y_values = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x_values, y_values)
Z = radial_basis_function(np.stack([X, Y], axis=-1), c, beta)

# Plot the RBF
plt.figure(figsize=(8, 6))
contour = plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(contour)
plt.title("Radial Basis Function")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Evaluate the radial basis function along x_values
rbf_1d = radial_basis_function(x_values[:, np.newaxis], c, beta)

# Plot the 1D radial basis function
plt.figure(figsize=(8, 6))
plt.plot(x_values, rbf_1d, label="RBF (1D)", color="blue")
plt.title("Radial Basis Function (1D)")
plt.xlabel("X-axis")
plt.ylabel("RBF Value")
plt.legend()
plt.grid(True)
plt.show()
```
